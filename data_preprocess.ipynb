## 0. 将图片按wsi分类

```python
import torch
import os
import numpy as np
folder_path = 'C:/Users/86136/Desktop/oold/mae_gnn_data/breast4'
save_dict_path = 'C:/Users/86136/Desktop/new_gcn/wsi_dict.npy'
save_dict = {}
folder_clas = os.listdir(folder_path)
print(folder_clas)
for clas in folder_clas:
    img_path = os.listdir(os.path.join(folder_path, clas))
    for img in img_path:
        wsi_num = img.split("_")[0]
        if str(wsi_num) not in save_dict.keys():
            save_dict[str(wsi_num)] = []
            save_dict[str(wsi_num)].append(img)
            continue
        save_dict[str(wsi_num)].append(img)
# np.save(save_dict_path, save_dict)
```

```python
wsi_dict = np.load(save_dict_path, allow_pickle='TRUE')
wsi = wsi_dict.item()['43']
a = [int(i.split("_")[4]) for i in wsi]
a.sort()
print(wsi)
```
